```python
from ultralytics import YOLO

model = YOLO("yolo12s.pt")

model.info()
```

```python
model.train(
    data="data/data.yaml",  # Dataset configuration
    epochs=100,               # Number of epochs
    patience=25,              # Early stopping patience
    imgsz=640,                # Input image size
    batch=8,                  # Batch size
    lr0=0.005,                # Initial learning rate
    device=0,            # Device (auto-select CUDA or CPU)
    project="training-iteration",   # Project directory
    name="yolo12s",      # Run name
    save=True,                # Save visualizations/checkpoints
    save_period=5,            # Save every N epochs

    # Data augmentation
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=0.0,
    translate=0.1,
    scale=0.5,
    shear=0.0,
    perspective=0.0,
    flipud=0.0,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.1,
    copy_paste=0.1,
    close_mosaic=10,
)
```

```python
model1 = YOLO("training-iteration/yolo12s/weights/best.pt")
model1.predict(
    source="data/test/images",  # Test images directory
    conf=0.45,     # Confidence threshold
    batch=16,      # Inference batch size
    save=True      # Save visualization results to runs/detect/predict
)
```

```python
metrics = model1.val()  # Evaluate the current weights and return metrics
print(metrics)
```

```python
import csv

# List of models and their paths
models = [
    ("YOLOv8s", "training-iteration/yolov8s/weights/best.pt"),
    ("YOLOv11s", "training-iteration/yolo11s/weights/best.pt"),
    ("YOLOv12s", "training-iteration/yolo12s/weights/best.pt")
]

# Prepare data for CSV
csv_data = []
headers = ["Model", "mAP@0.5", "mAP@0.5:0.95", "Precision", "Recall", "Fitness"]

for model_name, model_path in models:
    model = YOLO(model_path)
    metrics = model.val()
    
    # Extract key metrics
    row = [
        model_name,
        metrics.box.map50,  # mAP@0.5
        metrics.box.map,    # mAP@0.5:0.95
        metrics.box.mp,     # Mean Precision
        metrics.box.mr,     # Mean Recall
        metrics.fitness     # Fitness score
    ]
    csv_data.append(row)

# Write to CSV
with open("validation_results.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(headers)
    writer.writerows(csv_data)

print("Validation results exported to validation_results.csv")
```
